The dataset we're going to use is on this [link]("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/violencia_rio.csv"), it's a `csv` file.

Note: You can also see an interactive dashboard of violence in *Rio de Janeiro* in the following link:

* http://www.ispvisualizacao.rj.gov.br/index.html

### Statistical information on violence in *Rio de Janeiro*

The violence happens everywhere, there are no minimum requisites to this happen, even in the state of the "[*cidade maravilhosa*]("https://www.youtube.com/watch?v=Pq1LzATB-RM)" it happens a lot.

Continuing our analysis, we're going to group our dataset by year of occurrence (`vano`), then remove the `mes` column, and finally, we're going to use the method `describe` to generate some descriptive statistics.

```
# grouping by `vano` variable
rio = rj.groupby('vano').sum()

# removing the `mes` column
rio.drop(columns=['mes'], inplace=True)

# verifyin the first 5 inputs
rio.head()
```

```
# generating statistics
rio.describe()
```

We have a lot of information right here. For example:

* The maximum number of intentional homicides (`hom_doloso`) in a year is almost 8500 and the minimum is about 2700. The average of intentional homicides is about 6000 a year, is 500 a month, or 16 a day. That means that, on average, sixteen people are murdered in *Rio de Janeiro* everyday.

* The average number of rapes a year is about 2500, that is, every year this number of women are raped. It's almost 7 women a day. 

Looking closer in these informations, we can note some outliers, for example in the `roubo_veiculo` variable. As we can verify:



```
# describe roubo_veiculo column
rio.roubo_veiculo.describe()
```

```
# plotting a boxplot to verify outliers
rio.roubo_veiculo.plot(kind='box');
```

```
# removing outliers
rio_clean = rio.copy()
rio_clean.drop(rio_clean[rio_clean.roubo_veiculo > 40000].index, axis=0, inplace=True)

# describing `roubo_veiculo` of the new dataframe
rio_clean.roubo_veiculo.describe()
```

```
# plotting a bloxplot of `roubo_veiculo` from `rio_clean`

rio_clean.roubo_veiculo.plot(kind='box')
```

Now, we can see more about burglary<sup>1</sup> and robbery vehicles:



<sub> 1 Burglary is when someone takes something from someone else without its permission. Robbery is the same thing, but it happens with the use of a threat or firearm that reduces the victim's ability to defend or resist. </sub>

```
# average number of robbery vehicles
a = rio_clean.roubo_veiculo.mean()

# average number of burglary vehicles
b = rio_clean.furto_veiculos.mean()

# average number of recovered vehicles
c = rio_clean.recuperacao_veiculos.mean()

print("Robbery vehicles a year: {:.0f}".format(a))
print("Burglary vehicles a year: {:.0f}".format(b))
print("Recovered vehicles a year: {:.0f}".format(c))
```

How can we see the percentage of recovered vehicles related to robbery and burglary vehicles?

Let's check it out:

First, we must know that we'll find this value by this following expression:

  <center>$\frac{\text{recuperacao_veiculos}}{\text{roubo_veiculo} + \text{furto_veiculos}}$</center>

Then, let's code:

```
# percentage of recovered vehicles related to robbery and burglary vehicles
x = (rio_clean.recuperacao_veiculos.sum()/(rio_clean.roubo_veiculo.sum() + rio_clean.furto_veiculos.sum()))*100
print('{:.0f}% of all vehicles are recovered'.format(x))
```

In *Rio de Janeiro*, 41%  of all robbery and burglary vehicles are recovered by the Military Police.
